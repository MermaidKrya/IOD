```python
import re
from tqdm import tqdm
import pandas as pd
```

```python
with open("new.txt", encoding="utf-8") as newsfile: # Файл с новостями.
    text_news = [(n.split("-----\n")[0].split('\n')[0], 
                  n.split("-----\n")[0].split('\n')[1], 
                  n.split("-----\n")[1]) for n in newsfile.read().split("=====\n")[1:]]
    news = pd.DataFrame(text_news, columns = ['Header', 'Date', 'News'])
```

```python
strange_words = re.findall('[A-Za-z]+[ä-ü]+[A-Za-z]+', news.News.iloc[0]) #только с диакритическими знаками
strange_words
```
