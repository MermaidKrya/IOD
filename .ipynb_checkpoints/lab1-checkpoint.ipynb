```python
import numpy as np
from scipy import ndimage
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

```python
def normalize (array):
    return array / np.amax(array)     
```

```python
k = np.ones((3, 3))
k *= -1
for i in range(3):
    k[i,1] = 3
```

```python
image = Image.open("2.bmp")
image = 255 - np.array(image)[:, :, 0]
image = image % 2

without_verticals = normalize(ndimage.convolve(image, k, mode = "constant", cval = 0))
```

```python
_, axs = plt.subplots(1, 2, figsize = (20, 20))
axs[0].imshow(image, cmap = "gray")
axs[1].imshow(without_verticals, cmap = "gray")
```
